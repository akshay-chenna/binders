# Analysis of hcab64 

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_csv('A1.sc', delim_whitespace=True)
df2 = pd.read_csv('A2.sc', delim_whitespace=True)
df3 = pd.read_csv('A3.sc', delim_whitespace=True)
df4 = pd.read_csv('A4.sc', delim_whitespace=True)
df5 = pd.read_csv('C2.sc', delim_whitespace=True)
df6 = pd.read_csv('C4.sc', delim_whitespace=True)
```

```python
plt.figure(figsize=(8,6))
plt.rcParams['font.size'] = '16'
x = 'pae_interaction'
sns.kdeplot(df1[x])
sns.kdeplot(df2[x])
sns.kdeplot(df3[x])
sns.kdeplot(df4[x])
sns.kdeplot(df5[x])
sns.kdeplot(df6[x])
plt.xlabel('pAE interaction')
plt.legend(['A1','A2','A3','A4','C2','C4'])
plt.savefig('pAE_interaction.png', dpi=600, bbox_inches='tight',facecolor='white')
```

```python
plt.figure(figsize=(8,6))
plt.rcParams['font.size'] = '16'
x = 'target_aligned_rmsd'
sns.kdeplot(df1[x])
sns.kdeplot(df2[x])
sns.kdeplot(df3[x])
sns.kdeplot(df4[x])
sns.kdeplot(df5[x])
sns.kdeplot(df6[x])
plt.xlabel('Target aligned RMSD')
plt.legend(['A1','A2','A3','A4','C2','C4'])
plt.savefig('target_aligned_rmsd.png', dpi=600, bbox_inches='tight',facecolor='white')
```

```python
plt.figure(figsize=(8,6))
plt.rcParams['font.size'] = '16'
x = 'plddt_binder'
sns.kdeplot(df1[x])
sns.kdeplot(df2[x])
sns.kdeplot(df3[x])
sns.kdeplot(df4[x])
sns.kdeplot(df5[x])
sns.kdeplot(df6[x])
plt.xlabel('pLDDT')
plt.legend(['A1','A2','A3','A4','C2','C4'])
plt.savefig('pLDDT_binderpng', dpi=600, bbox_inches='tight',facecolor='white')
```
